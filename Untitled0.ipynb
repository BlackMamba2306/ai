```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np
```

```python
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
train_images, test_images = train_images / 255.0, test_images / 255.0

class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

classes_to_keep = ['airplane', 'automobile', 'cat', 'deer', 'dog']
indices_to_keep = [class_names.index(c) for c in classes_to_keep]
```

```python
train_mask = np.isin(train_labels, indices_to_keep).flatten()
train_images = train_images[train_mask]
train_labels = train_labels[train_mask]

test_mask = np.isin(test_labels, indices_to_keep).flatten()
test_images = test_images[test_mask]
test_labels = test_labels[test_mask]

old_to_new = {old: new for new, old in enumerate(indices_to_keep)}
train_labels = np.array([old_to_new[l[0]] for l in train_labels])
test_labels = np.array([old_to_new[l[0]] for l in test_labels])

class_names = classes_to_keep

print(f"New class_names: {class_names}")
print(f"Train images: {train_images.shape}, Train labels: {train_labels.shape}")
print(f"Test images: {test_images.shape}, Test labels: {test_labels.shape}")

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(len(class_names), activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()

history = model.fit(train_images, train_labels, epochs=20, batch_size=64,
                    validation_data=(test_images, test_labels))

test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print(f"\nTest accuracy: {test_acc}")
```
